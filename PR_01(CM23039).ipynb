```python
import numpy as np
import pandas as pd

from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean, cityblock
import Levenshtein
```

```python
customers = pd.DataFrame({
    'Customer':['A','B','C','D'],
    'Age':[25,28,35,24],
    'Annual Income':[40000,45000,80000,42000],
    'Spending Score':[80,78,40,85]
})

customers
```

```python
customer1 = customers.iloc[0,1:].values
customer2 = customers.iloc[1,1:].values

print("Customer A:", customer1)
print("Customer B:", customer2)
```

```python
distance = euclidean(customer1, customer2)

print("Euclidean Distance =", distance)
```

```python
distance = cityblock(customer1, customer2)

print("Manhattan Distance =", distance)
```

```python
similarity = cosine_similarity([customer1],[customer2])

print("Cosine Similarity =")
print(similarity)
```

```python
text1 = "Laptop"
text2 = "Laptops"

distance = Levenshtein.distance(text1,text2)

print("Edit Distance =",distance)
```

```python
target = customers.iloc[0,1:].values

print("Similarity with Customer A\n")

for i in range(len(customers)):

    current = customers.iloc[i,1:].values

    euc = euclidean(target,current)
    man = cityblock(target,current)
    cos = cosine_similarity([target],[current])[0][0]

    print("Customer",customers.iloc[i,0])
    print("Euclidean Distance :",round(euc,2))
    print("Manhattan Distance :",round(man,2))
    print("Cosine Similarity :",round(cos,4))
    print("----------------------------")
```

```python
print("Customer B is most similar to Customer A because:")
print("- Lowest Euclidean Distance")
print("- Lowest Manhattan Distance")
print("- Highest Cosine Similarity")
```
